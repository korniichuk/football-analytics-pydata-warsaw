### **Sportmonks: Fixture (match) events**

```python
import os

import pandas as pd
import requests
from dotenv import load_dotenv
```

```python
load_dotenv(".env")

API_TOKEN = os.getenv("API_TOKEN")
params = {"api_token": API_TOKEN, "include": "events"}
```

```python
data = requests.get("https://api.sportmonks.com/v3/football/fixtures/19427866", params=params).json()["data"]
match = pd.json_normalize(data)
```

```python
events = pd.json_normalize(match.events[0])
len(events)
```
