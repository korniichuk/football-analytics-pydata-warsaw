### **Sportmonks: Fixture (match) statistics**

```python
import os

import pandas as pd
import requests
from dotenv import load_dotenv
```

```python
load_dotenv(".env")

API_TOKEN = os.getenv("API_TOKEN")
params = {"api_token": API_TOKEN, "include": "statistics.type;events;formations;timeline;lineups.details.type"}
```

```python
data = requests.get("https://api.sportmonks.com/v3/football/fixtures/19427866", params=params).json()["data"]
match = pd.json_normalize(data)
match.columns
```

```python
statistics = pd.json_normalize(match.statistics[0])
```
