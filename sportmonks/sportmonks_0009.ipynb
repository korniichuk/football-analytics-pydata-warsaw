### **Sportmonks: Fixture (match) lineups**

```python
import os

import pandas as pd
import requests
from dotenv import load_dotenv
```

```python
load_dotenv(".env")

API_TOKEN = os.getenv("API_TOKEN")
params = {"api_token": API_TOKEN, "include": "lineups.details.type"}
```

```python
data = requests.get("https://api.sportmonks.com/v3/football/fixtures/19427866", params=params).json()["data"]
match = pd.json_normalize(data)
```

```python
lineups = pd.json_normalize(match.lineups[0])
len(lineups)
```

```python
def head_skip_all_nan(df, n):
    return df.dropna(axis=1, how='all').head(n)
    
head_skip_all_nan(lineups.drop('details', axis=1), 10)
```

```python
lineups.to_pickle("lineups.pkl")
```
