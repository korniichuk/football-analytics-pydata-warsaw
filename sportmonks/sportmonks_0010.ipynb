### **Sportmonks: Fixture (match) lineups details**

```python
import pandas as pd
import requests
```

```python
lineups = pd.read_pickle("lineups.pkl")
```

```python
lineups_filtered = lineups[lineups['details'].apply(len) > 0]

exploded = lineups_filtered.explode('details').reset_index(drop=True)
normalized = pd.json_normalize(exploded["details"])
pd.concat([exploded.drop('details', axis=1), normalized[["data.value", "type.name"]]], axis=1)
```
